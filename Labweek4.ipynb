```python
sum_even = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    if num % 2 == 0:
        sum_even += num

print("Final sum of even positive numbers:", sum_even)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):
    product = num * i

    if product > 50:
        break

    if product % 3 == 0:
        continue

    print(num, "x", i, "=", product)
```

```python
import random

secret = random.randint(1, 20)
attempts = 0

while True:
    guess = int(input("Guess the number (1 to 20): "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1

    if guess == secret:
        print("Correct! Number of valid attempts:", attempts)
        break
```

```python
n = int(input("Enter a number n: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count of numbers divisible by 4:", count)
```

```python
for i in range(1, 6):

    if i * i > 6:
        break

    for j in range(1, i + 1):

        if j == 3:
            continue

        print(j, end=" ")

    print()
```
